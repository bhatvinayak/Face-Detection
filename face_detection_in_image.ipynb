```python
import cv2
```

```python
im = imread("/Users/B G Vinayak/Desktop/test.jpg")
gray = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
title("Original Image")
axis("off")
imshow(im)
show()

title("Grayscale Image")
axis("off")
imshow(gray, cmap= pylab.gray())
show()
```

```python
cascadePath = "/Users/B G Vinayak/Desktop/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascadePath)
faces = faceCascade.detectMultiScale(im)
print ("Number of faces found ->", faces.shape[0])
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(im, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
title("Image with faces detected")
axis("off")
imshow(im)
show()
```
